## agregate

```python
import pandas as pd
import numpy as np
```

```python
df1=pd.DataFrame({'grup':['A','B','C'],'degisken1':[10,2,3],
                 'degisken2':[234,56,36]},
               columns=['grup','degisken1','degisken2'])
```

```python
df1
```

```python
#aggregate
```

```python
df1.groupby("grup").mean()
```

```python
df1.groupby("grup").aggregate(["min",np.median,max])  #pandasta min var "" aldık
```

## Filter

```python
df=pd.DataFrame({'grup':['A','B','C'],'degisken1':[10,23,31],
                 'degisken2':[24,56,3]},
               columns=['grup','degisken1','degisken2'])
df
```

```python
def filtrele(x):
    return x["degisken1"].std()>9

#degisken1 e gore st spmsı 9dan buyuk olanları al dedik 
    
```

```python
df1.groupby("grup").std()
```

```python

df1.groupby("grup").filter(filtrele)
```

# Pivot

```python
import pandas as pd
import seaborn as sns
```

```python
titanic=sns.load_dataset('titanic')
```

```python
titanic.head()
```

```python
titanic.groupby("sex")["survived"].mean() #cinsiyete gore gruplayıp hayatta kalanlar
```

```python
titanic.groupby(["class","sex"])[["survived"]].aggregate("mean").unstack()
```

```python
#unstack kaldırılınca yapı zor okunuyor
```

```python
#pivot
```

```python
titanic.pivot_table("survived",index="sex",columns="class")
```

```python
titanic.age.head()
```

```python
age=pd.cut(titanic["age"],[0,18,90]) #yas aralıklarını bol
age.head(10)
```

```python
titanic.pivot_table("survived",["sex",age],"class")  #kategorik aldık ageyi
```
